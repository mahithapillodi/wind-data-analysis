```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
directory_path = '/Users/mpillodi/documents/GitHub/wind-data-analysis/csvfolder'
```

```python
wind_data = pd.DataFrame()
```

```python
for filename in os.listdir(directory_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory_path, filename)
        df = pd.read_csv(file_path)
        wind_data = pd.concat([wind_data, df], ignore_index=True)
```

```python
print(wind_data.head())
```

```python
statistics = wind_data.describe()
print(statistics)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(wind_data['s1a'], bins=20, kde=True)
plt.title('Distribution of Wind Speed 1 mean')
plt.xlabel('Wind Speed 1 mean (m/s)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.set(style="whitegrid")

for column in wind_data.columns:
    if column not in ['date', 'time', 'period']:
        plt.figure(figsize=(8, 6))
        
        if wind_data[column].dtype in ['int64', 'float64']:
            sns.histplot(wind_data[column], bins=20, kde=True)
            plt.title(f'Distribution of {column}')
            plt.xlabel(column)
            plt.ylabel('Frequency')
        
        else:
            sns.countplot(data=wind_data, x=column)
            plt.title(f'Count of {column}')
            plt.xlabel(column)
            plt.ylabel('Count')
            plt.xticks(rotation=45, ha='right')
        plt.show()
```
